```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#Line Plot

```python
x = [1,2,3,4,5]
y = [10,14,8,16,12]

plt.plot(x,y, marker = 'x', linestyle = '-.', color = 'r', label = 'Line')

plt.title('Line Plot Example')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.grid(True)
plt.show()
```

#Scatter Plot

```python
x = [1,2,3,4,5]
y = [10,14,8,16,12]

plt.scatter(x,y, marker = 'v', color = 'b', label = 'Data Points')

plt.title('Scatter Plot Example')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.grid(False)
plt.show()
```

#Bar Plot

```python
categories = ['A','B','C','D']
values = [15,8,12,10]

plt.bar(categories, values, color = 'g', alpha = 0.5) #alpha --> adjusts the transparency of the bars
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis = 'y')
plt.show()
```

#Histogram

```python
data = [1,2,2,3,3,3,4,4,5,5,5,5]

plt.hist(data, bins = 5, color='r', alpha=0.5)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='x')
plt.show()
```

#Pie Chart

```python
labels = ['A','B','C','D']
sizes = [20,15,25,40]

plt.pie(sizes,labels=labels, autopct = '%1.2f%%', startangle = 90, colors = ['gold','lightgreen',"blue","green"])
#autopct --> %1.1f%% --> displays the percentage of each category on the pie chart with one decimal place

plt.title('Pie Chart Example')
plt.axis("equal") # --> Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
df = pd.read_csv('data2.csv')
df
```

#Line Plot

```python
plt.plot(df['Name'], df['Age'], marker = 'D', linestyle = '-', color = 'b')

plt.title('Line Plot - Age vs Name')
plt.xlabel('Name')
plt.ylabel('Age')
plt.grid(True)
plt.xticks(rotation=120)
plt.show()
```

#Scatter Plot

```python
plt.scatter(df['Salary'], df['Rating'], marker = 'o', color = 'r', label = 'Data Points')

plt.title('Scatter Plot - Salary vs Rating')
plt.xlabel('Salary')
plt.ylabel('Rating')
plt.legend()
plt.grid(True)
plt.show()
```

#Bar Plot

```python
plt.figure(figsize = (8,6))

plt.bar(df['Name'], df['Age'], color = 'orange', alpha = 0.5)
plt.title('Bar Plot - Salary by Name')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
